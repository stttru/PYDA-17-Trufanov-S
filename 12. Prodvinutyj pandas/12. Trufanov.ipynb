## Задание 1

```python
import pandas as pd

log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
def define_source_type(row):
    if row.traffic_source == 'yandex' or row.traffic_source == 'google':
        return 'organic'
    elif row.traffic_source == 'paid' and row.region == 'Russia':
        return 'ad'
    elif row.traffic_source == 'paid' and row.region != 'Russia':
        return 'other'
    else:
        return row.traffic_source
    
log['source_type'] = log.apply(define_source_type, axis=1)

log.head(20)
```
