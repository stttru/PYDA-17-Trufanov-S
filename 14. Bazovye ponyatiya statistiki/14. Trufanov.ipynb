### Задание 2. Работа с выбросами

```python
#найдем выбросы через межкваритльный размах
bounds = dict()
for param in ['rectal temperature','pulse', 'respiratory rate', 'nasogastric reflux PH']:
    print('\t', '\033[1m', param.upper(), '\033[0m')
    q1 = data[param].quantile(0.25)
    q3 = data[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    bounds[param] = (lower_bound, upper_bound)
    print('Нижняя граница', round(lower_bound,2))
    print('Верхняя граница', round(upper_bound,2))
    print('Среднее значение с выбросами',data[param].mean())
    print('Среднее значение без выбросов',data[(data[param] > lower_bound) & (data[param] < upper_bound)][param].mean())
#     remove_outliers = remove_outliers[remove_outliers[param].between(lower_bound, upper_bound, inclusive=True)]
```

```python
bounds
```

с выбросами и без них практически не изменилась, поэтому данные выбросы оставляем.

## Пульс
* Верхняя граница пульса 148 уд/мин
* нормальные значения для взрослых лошадей 30 - 40 уд/мин
* нормальные значения для жеребцов 80 - 100 уд/мин
* усиленное сердцебиения может говорить о болезненных поражениях или о циркуляторном шоке, который испытывает лошадь
#### пробуем убрать выбросы

```python
remove_outliers = data[data['pulse'].between(bounds['pulse'][0], bounds['pulse'][1], inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
# Какие выбросы убрали:
outliers = pd.concat([data, remove_outliers]).drop_duplicates(keep=False)
print('Количество пропусков:', outliers['pulse'].isnull().sum())
outliers[outliers['pulse'] > 0]
```
