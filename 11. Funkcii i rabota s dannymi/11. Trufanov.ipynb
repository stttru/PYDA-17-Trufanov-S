## Задание 1

```python
import pandas as pd

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
def classify(param):
    if param <= 2:
        return 'низкий рейтинг'
    elif param <=4:
        return 'средний рейтинг'
    elif param == 4.5 or param == 5:
        return 'высокий рейтинг'
    else: 
        return 'рейтинг не определен'
    
ratings['class'] = ratings['rating'].apply(classify)
ratings
```

## Задание 2

```python
keywords = pd.read_csv('ml-latest-small/keywords.csv')
keywords.head()
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    
}

def geo(param):
    for item, val in geo_data.items():
        for city in val:
            if city in param:
                return item
            
keywords['region'] = keywords['keyword'].apply(geo)
keywords
```
