# рабочие данные - Credit Card customers

```python
df = pd.read_csv('BankChurners.csv')
df
```

```python
df.drop(df.columns[[-1, -2, -4, -7]], axis='columns', inplace = True)
```

```python
[column for column in df.columns]
```

```python
df.describe()
```

```python
Income_Category_Credit_Limit = df[['Income_Category', 'Credit_Limit']].groupby(by='Income_Category').mean()
Income_Category_Credit_Limit
```

```python
Income_Category_Credit_Limit.loc[['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown']].plot(kind='bar', 
                                                              title='Средний кредитный лимит в зависимости от годового дохода клиента')
```

### По визуализации можно сказать, что чем выше годовой доход клиента, тем выше его средний кредитный лимит. Это логично, так как лимит по кредиту устанавливается на основе платежеспособности клиента.

```python
plt.plot(df.Customer_Age, df.Months_on_book, 'o')
plt.title('Период взаимодействия с банком')
plt.xlabel('Возраст клиента')
plt.ylabel('Количество месяцев взаимодействия')
```

### Чем выше возраст клиента, тем выше количество месяцев взаимодействия его с банком

```python
Age_avg_creadit_limit = df[['Customer_Age', 'Credit_Limit']].groupby('Customer_Age').mean()
Age_avg_creadit_limit.head()
```

```python
plt.plot(Age_avg_creadit_limit.index, Age_avg_creadit_limit.Credit_Limit)
plt.title('Средняя сумма кредитного лимита по возрасту клиента')
plt.xlabel('Возраст клиента')
plt.ylabel('Средняя сумма кредитного лимита')
```

### Из визуализации средней суммы кредитного лимита в зависимости от возраста клиента видно, что клиенты в возрасте 50 лет имеют больший кредитный лимит (видимо, в этом возрасте больше состоятельных), есть пик кредитного лимита у молодых в районе 25 лет, и минимум около 35 лет, также с увеличением возраста от 60 лет сумма кредитного лимита падает.

```python
# отношение непогашенного остатка по кредитной карте к кредитному лимиту в зависимости от уровня образования
Avg_Utilization_Ratio_Education_Level = df[['Avg_Utilization_Ratio', 'Education_Level']].groupby('Education_Level').mean()
Avg_Utilization_Ratio_Education_Level
```

```python
Avg_Utilization_Ratio_Education_Level.loc[['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Unknown']].plot(kind='bar', 
                                                                                                                                              title='отношение непогашенного остатка по кредитной карте к кредитному лимиту'
                                                                                                                                            )
```

### Интересно заметить, что отношение среднего непогашенного остатка по кредитной карте к кредитному лимиту примерно одинаково для всех уровней образования клиентов. 

```python
# количество клиентов в зависимости от семейного положения
Number_clients_mar_status = df[['CLIENTNUM', 'Marital_Status']].groupby('Marital_Status').count()
Number_clients_mar_status
```

```python
Number_clients_mar_status.loc[['Single', 'Married', 'Divorced', 'Unknown']].plot(kind='bar', 
                                        title='Количество клиентов в зависимости от их семейного положения')
```

###  Больше всего клиентов в браке, далее идут одинокие, разведенных совсем мало, как и клиентов с неизвестным семейным положением.

```python
Age_Gender_Dependent_count = df.pivot_table(index = 'Customer_Age', columns = 'Gender', values = 'Dependent_count', aggfunc = 'mean', fill_value = 0)
Age_Gender_Dependent_count.head()
```

```python
Age_Gender_Dependent_count.plot(kind='bar')
plt.title('Среднее количество иждивенцев в зависимости от возраста и пола клиента')
plt.xlabel('Возраст клиента')
plt.ylabel('Количество иждивенцев')
```

###  По визуализации можно сказать, что количество иждивенцев мало зависит от пола клиента, больше всего иждивенцев (около 3) у клиентов в возрасте от 40 до 48.  Поровну (около 2) для клиентов от 33 до 39 и от 50 до 58. Данное распределение очень похоже на нормальное.
