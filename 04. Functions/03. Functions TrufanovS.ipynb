# Исходные данные

```python
# Исходные данные

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


# Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

def document_owner(documents, document_number):    
    for document in documents:
        if document['number'] == document_number:
            return document['name']
    return('Документ не найден в базе, возврат к выбору команд')

# Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def shelf_number(directories, document_number):
    for shelf, document in directories.items():
        for element in document:
            if element == document_number:
                return shelf
    return None


# Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам

def full_information(documents):
    for document in documents:
        print(f"№: {document['number']}, " +
              f"тип: {document['type']}, " + 
              f"владелец: {document['name']}, " + 
              f"полка хранения: {shelf_number(directories, document['number'])}")  
        
# Пункт 4. Пользователь по команде "ads" может добавить новую полку  

def add_shelf(directories, number_shelf):
    if number_shelf in directories:
        return(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}")
    else:
        directories[number_shelf] = []
        return(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}")

# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def delete_self(directories, number_self):
    if number_self in directories:
        if directories[number_self] == []:
            del directories[number_self]
            return(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}")
        else:
            return("На полке есть документы, удалите их перед удалением полки. " + 
                   f"Текущий перечень полок: {', '.join(directories.keys())}")
    else:
        return(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}")

print('''Доступные команды:
        p - узнать владельца документа по его номеру
        s - по номеру документа узнать на какой полке он хранится
        l - увидеть полную информацию по всем документам
        ads - добавить новую полку
        ds - удалить существующую полку из данных (только если она пустая)
        q - выход из программы''')

def input_comand():
    
    while True:
        user_input = input('Введите команду : ')
        if user_input == 'p':
            document_number = input('Введите номер документа: ')
            print(document_owner(documents, document_number))  
        elif user_input == 's':
            document_number = input('Введите номер документа: ')
            shelf_number(directories, document_number)
            if shelf_number(directories, document_number) is None:
                print('Документ не найден в базе')
            else:
                print(f'Документ хранится на полке: {shelf_number(directories, document_number)}')
        elif user_input == 'l':
            full_information(documents)
        elif user_input == 'ads':
            number_shelf = input('Введите номер полки которую добавить: ')
            print(add_shelf(directories, number_shelf))
        elif user_input == 'ds':
            number_self = input('Введите номер полки которую удалить: ')
            print(delete_self(directories, number_self))
        elif user_input == 'q':
            print('Завершение программы')
            break
        else: print("Некорректный ввод команды, повторите ввод команды согласно списку доступных комманд")

input_comand()
```
